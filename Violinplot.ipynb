# Violinplot using Seaborn in Python
### Basic visualization of “fmri” dataset using violinplot 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
	
sns.set(style = 'whitegrid')
fmri = sns.load_dataset("fmri")
	
sns.violinplot(x ="timepoint",
			y ="signal",
			data = fmri)
```

```python

sns.set(style = 'whitegrid')
fmri = sns.load_dataset("fmri")
	
sns.violinplot(x ="timepoint",
			y ="signal",
			hue ="region",
			style ="event",
			data = fmri)
```

```python
sns.set(style = 'whitegrid')
tip = sns.load_dataset('tips')

sns.violinplot(x ='day', y ='tip', data = tip)
```

```python

sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.violinplot(x=tip["total_bill"])
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.violinplot(x="tip", y="day", data=tip)
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.violinplot(x="day", y="total_bill", hue="time", data=tips)
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.violinplot(x ='day', y ='tip', data = tip, linewidth = 4)
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.violinplot(x="day", y="total_bill", hue="smoker",
				data=tips, palette="Set2", dodge=True)
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.violinplot(x="time", y="tip", data=tips,
				order=["Dinner", "Lunch"])
```

```python

# use to set style of background of plot
sns.set(style="whitegrid")

# loading data-set
tips = sns.load_dataset("tips")

sns.violinplot(x="day", y="total_bill", hue="sex",
					data=tip, palette="Set2", split=True,
					scale="count")
```

```python

datasets = sns.load_dataset('iris')

datasets.head()

sns.violinplot(x='species', y='sepal_width', data=datasets,
			inner="quart", linewidth=1)
```
